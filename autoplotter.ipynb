# **Import libraries**

```python
from autoplotter import run_app
import pandas as pd
```

# **Loading data**

```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv')
```

```python
df
```

# **Run App**

```python
run_app(df,mode = "inline")
```

```python
run_app(df,mode='external')
```
